```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
```

```python
lin_reg = LinearRegression()
tree_reg = DecisionTreeRegressor(random_state = 42)
rf_reg = RandomForestRegressor(random_state = 42)
```

```python
air_prepared.shape, air_label.shape
```

```python
# LinearRegression 
lin_scores = cross_val_score(lin_reg, air_prepared, air_label, scoring = 'neg_mean_squared_error', cv = 10, n_jobs = -1)
lin_rmse = np.sqrt(-lin_scores.mean())
lin_rmse
```

```python
# DecisionTree 
tree_scores = cross_val_score(tree_reg, air_prepared, air_label,scoring = 'neg_mean_squared_error', cv = 10, n_jobs = -1)
tree_rmse = np.sqrt(-tree_scores.mean())
tree_rmse
```

```python
# Randomforest 
rf_scores = cross_val_score(rf_reg, air_prepared, air_label,scoring = 'neg_mean_squared_error', cv = 10, n_jobs = -1)
rf_rmse = np.sqrt(-rf_scores.mean())
rf_rmse
```

###   

```python
#  
from sklearn.model_selection import GridSearchCV

rf_reg = RandomForestRegressor(random_state = 42)

param_grid = {'n_estimators' : [30, 50, 100], 'max_features' : [2, 4, 6, 8]}

grid_search = GridSearchCV(rf_reg, param_grid, scoring = 'neg_mean_squared_error', cv = 5, n_jobs = -1)

grid_search.fit(air_prepared, air_label)
```

```python
from sklearn import set_config
set_config(display='diagram')
```

```python
grid_search.best_params_
```

```python
grid_search.best_estimator_
```

```python
cv_results = grid_search.cv_results_
for mean_score, params in zip(cv_results['mean_test_score'], cv_results['params']):
    print(np.sqrt(-mean_score), params)
```

```python
#  
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint

param_distribs = {'n_estimators' : randint(low = 1, high = 200),
                  'max_features' : randint(low = 1, high = 8)
                 }

rnd_search = RandomizedSearchCV(rf_reg, param_distribs, n_iter = 10, scoring = 'neg_mean_squared_error', cv = 5, n_jobs = -1, random_state = 42)
rnd_search.fit(air_prepared, air_label)
```

```python
rnd_search.best_params_
```

```python
rnd_search.best_estimator_
```

```python
cv_results = rnd_search.cv_results_
for mean_score, params in zip(cv_results['mean_test_score'], cv_results['params']):
    print(np.sqrt(-mean_score), params)
```
