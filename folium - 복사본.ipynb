```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn

df = pd.read_csv("./datasets/train.csv")
```

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(20, 12))

sns.scatterplot(data = df[ df['city'] == 'Boston'  ], x = 'longitude',y= 'latitude',hue = 'room_type', ax=axes[0][2])
sns.scatterplot(data = df[ df['city'] == 'Chicago'  ], x = 'longitude',y= 'latitude',hue = 'room_type', ax=axes[0][1])
sns.scatterplot(data = df[ df['city'] == 'DC'  ], x = 'longitude',y= 'latitude',hue = 'room_type', ax=axes[1][1])
sns.scatterplot(data = df[ df['city'] == 'LA'  ], x = 'longitude',y= 'latitude',hue = 'room_type', ax=axes[1][0])
sns.scatterplot(data = df[ df['city'] == 'NYC'  ], x = 'longitude',y= 'latitude',hue = 'room_type', ax=axes[1][2])
sns.scatterplot(data = df[ df['city'] == 'SF'  ], x = 'longitude',y= 'latitude',hue = 'room_type', ax=axes[0][0])


ax=axes[0][2].set_title('Boston')
ax=axes[0][1].set_title('Chicago')
ax=axes[1][1].set_title('DC')
ax=axes[1][0].set_title('LA')
ax=axes[1][2].set_title('NYC')
ax=axes[0][0].set_title('SF')
```

```python
import folium as g
```

```python
NYC_DF = df[ df['city'] == 'NYC']
NYC_DF1 = NYC_DF[ NYC_DF['room_type'] == 'Entire home/apt']
NYC_DF2 = NYC_DF[ NYC_DF['room_type'] == 'Private room']
NYC_DF3 = NYC_DF[ NYC_DF['room_type'] == 'Shared room']

nyc_w1 = NYC_DF1[['longitude','latitude' ] ]
nyc_w2 = NYC_DF2[['longitude','latitude' ] ]
nyc_w3 = NYC_DF3[['longitude','latitude' ] ]
nyc_map = g.Map(location = [ nyc_w1.iloc[0][1] , nyc_w1.iloc[0][0] ],
zoom_start = 12)
nyc_df = df[ df['city'] == 'NYC'  ]

for i in range(len( NYC_DF1 )):
        g.Circle(
        location = [ nyc_w1.iloc[i][1] , nyc_w1.iloc[i][0] ],
        radius = 20,
        color = 'red').add_to(nyc_map)
        
for i in range(len( NYC_DF2 )):
        g.Circle(
        location = [ nyc_w2.iloc[i][1] , nyc_w2.iloc[i][0] ],
        radius = 20,
        color = 'blue').add_to(nyc_map)
        
for i in range(len( NYC_DF3 )):
        g.Circle(
        location = [ nyc_w3.iloc[i][1] , nyc_w3.iloc[i][0] ],
        radius = 20,
        color = 'green').add_to(nyc_map)
        

nyc_map
```
